```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import warnings
import random
import plotly.express as px
from sklearn import datasets
print("Libraries imported")
```

```python
#data sets inder seaborn library
print(sns.get_dataset_names())
print(len(sns.get_dataset_names()))
```

```python
df_titanic=sns.load_dataset('titanic')
df_titanic.sample(2)
```

```python
# sns.countplot
# sns.barplot
# sns.boxplot
# sns.scatterplot
# sns.catplot
# sns.kdeplot
```

```python
#To count the values for one column based on another column
sns.countplot(data=df_titanic, x='sex', hue='alive', palette=sns.color_palette("viridis")) #hue: divide further based on categorical data
plt.title("Titanic survival based on Gender")
plt.show() #This code show us the anotamy, it will provide the lable by it's own
```

```python
sns.barplot(data=df_titanic, x='class', y='fare', hue='alive', palette=sns.color_palette('ocean',2)) #use estimator='mean' to change confidence interval
plt.show
#The bar is a confidence interval (fluctuation) showing the fare price average fluctuation 
#CI default value is 95%
```

```python
#To remove confidence interval use: use errorbar=('ci',0), ci value will be 0 (not calling any confidence interval)
sns.barplot(data=df_titanic, x='class', y='fare', hue='alive',errorbar=('ci',0), palette=sns.color_palette('ocean',2)) #use estimator='mean' to change confidence interval
plt.grid(color='b',alpha=0.1)
plt.show
```

```python
#sns.boxplot() to check outliers
sns.boxplot(data=df_titanic, x='class', y='fare', hue='sex')
plt.show()
```

```python
#sns.boxplot() to check skewness in outlirs
plt.figure(figsize=(10,5))
sns.boxplot(data=df_titanic, y='class', x='fare', hue='sex', orient='h')
plt.show()
```

```python
# sns.violinplot
plt.figure(figsize=(10,5))
sns.violinplot(data=df_titanic, x='class', y='fare', hue='sex', orient='v')
plt.show()
```
